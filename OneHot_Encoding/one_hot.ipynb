1. OneHotEncoding using Pandas

```python
df = pd.get_dummies(df, columns=['fuel','owner'])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df[['fuel_CNG'	,'fuel_Diesel','fuel_LPG','fuel_Petrol','owner_First Owner','owner_Fourth & Above Owner','owner_Second Owner','owner_Test Drive Car','owner_Third Owner']] = df[['fuel_CNG'	,'fuel_Diesel','fuel_LPG','fuel_Petrol','owner_First Owner','owner_Fourth & Above Owner','owner_Second Owner','owner_Test Drive Car','owner_Third Owner']].apply(LabelEncoder().fit_transform)
```
