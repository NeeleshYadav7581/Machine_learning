```python
import pandas as pd 
import requests
```

Response checking

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=42de417fe73183298a002e2b97a7a4a3&language=en-US&page=1')
```

Creating the dataset

```python
df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
df
```

Create a empaty dataset

```python
new_df = pd.DataFrame()
new_df
```

Create a loop for all pages 

```python
for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=42de417fe73183298a002e2b97a7a4a3&language=en-US&page={}'.format(i))
    df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    new_df = pd.concat([new_df, df], ignore_index=True)
```
